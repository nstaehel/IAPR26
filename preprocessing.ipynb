```python
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from preprocessing_fcts import *
```

```python
def load_image():
    # Define path
    id = np.random.randint(770, 983)
    print(id)
    path_he = f"images/train_images/L1000{id}.jpg"
    # Check if folder and image exist
    assert os.path.exists(path_he), "Image not found, please check directory structure"

    # Load image
    img_he = np.array(Image.open(path_he))

    # Display image
    plt.imshow(img_he)
    plt.axis('off')
    plt.tight_layout()

    return img_he
```

```python
img = load_image()


img_thresh_yellow, img_thresh_red, img_thresh_blue = preprocessing(img)

plt.imshow(img)

plt.imshow(img_thresh_blue)
plt.axis('off')
plt.tight_layout()
```
