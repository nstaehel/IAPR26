```python
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from preprocessing_fcts import *
```

```python
def load_image():
    # Define path
    id = np.random.randint(770, 983)
    print(id)
    path_he = f"images/train_images/L1000{id}.jpg"
    # Check if folder and image exist
    assert os.path.exists(path_he), "Image not found, please check directory structure"

    # Load image
    img_he = np.array(Image.open(path_he))

    # Display image
    # plt.imshow(img_he)
    # plt.axis('off')
    # plt.tight_layout()

    return img_he
```

```python
img = load_image()


img_thresh_yellow, img_thresh_red, img_thresh_blue, img_thresh_green, img_thresh_black = preprocessing(img)

# plt.imshow(img)

# plt.imshow(img_thresh_black)
# plt.axis('off')
# plt.tight_layout()

plt.figure(figsize=(15, 15))


# Add the first image to the figure (top-left position)
plt.subplot(3, 2, 1)  # 2 rows, 2 columns, first position
plt.imshow(img)  
plt.axis('off')  # Hide the axis labels
plt.title("Image") 

# Add the second image to the figure (top-right position)
plt.subplot(3, 2, 2)  # 2 rows, 2 columns, second position
plt.imshow(img_thresh_yellow)  
plt.axis('off')  # Hide the axis labels
plt.title("Image yellow") 

# Add the third image to the figure (bottom-left position)
plt.subplot(3, 2, 3)  # 2 rows, 2 columns, third position
plt.imshow(img_thresh_red) 
plt.axis('off')  # Hide the axis labels
plt.title("Image red")  

# Add the fourth image to the figure (bottom-right position)
plt.subplot(3, 2, 4)  # 2 rows, 2 columns, fourth position
plt.imshow(img_thresh_blue)  
plt.axis('off')  # Hide the axis labels
plt.title("Image blue")  

# Add the fourth image to the figure (bottom-right position)
plt.subplot(3, 2, 5)  # 2 rows, 2 columns, fourth position
plt.imshow(img_thresh_green)  
plt.axis('off')  # Hide the axis labels
plt.title("Image green")  

# Add the fourth image to the figure (bottom-right position)
plt.subplot(3, 2, 6)  # 2 rows, 2 columns, fourth position
plt.imshow(img_thresh_black)  
plt.axis('off')  # Hide the axis labels
plt.title("Image black")  

plt.show()
```
