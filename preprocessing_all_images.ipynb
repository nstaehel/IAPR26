```python
import os
from tqdm import tqdm
from PIL import Image

from image_utils import *
from preprocessing import *
```

```python
img_dir_train = os.path.join("iapr-26-uno-vision-challenge", "train_images")
img_dir_test = os.path.join("iapr-26-uno-vision-challenge", "test_images")
img_dir_ref = os.path.join("iapr-26-uno-vision-challenge", "reference_images")

img_dir_prepro_train = os.path.join("preprocessed_images", "train_images")
img_dir_prepro_test = os.path.join("preprocessed_images", "test_images")
img_dir_prepro_ref = os.path.join("preprocessed_images", "reference_images")
```

```python
background_img = generate_backgound_image(img_dir_train)
```

```python
def store_preprocessed_img(input_dir, output_dir, background_img):
    if (not os.path.exists(output_dir)):
        os.makedirs(output_dir)
        for image_file in tqdm(os.listdir(input_dir)):
            input_img_path = os.path.join(input_dir, image_file)
            output_img_path = os.path.join(output_dir, image_file)
            img = load_image(input_img_path, display=False)
            prepro_arr = remove_background(img, background_img)
            prepro_img = Image.fromarray(prepro_arr)
            prepro_img.save(output_img_path)
```

```python
store_preprocessed_img(img_dir_train, img_dir_prepro_train, background_img)
```

```python
store_preprocessed_img(img_dir_test, img_dir_prepro_test, background_img)
```

```python
store_preprocessed_img(img_dir_ref, img_dir_prepro_ref, background_img)
```
